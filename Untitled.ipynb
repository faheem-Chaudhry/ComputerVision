```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv



def convolution(image,mask):
                        #a output variable to hold results.
    output = np.zeros(shape=(image.shape[0], image. shape [1]))
                    #flipping mask horizantally
    flipMaskHorizantal=np.fliplr(mask)
                    #flipping mask vertically
    flipMaskVertical=np.flipud(flipMaskHorizantal)
                    #now flipped Mask is equal to Kernal
    kernel=flipMaskVertical
                    #getting shape or x/y coordinates of image
    (ix, iy) = image.shape[:2]
                    #getting shape or x/y coordinates of the kernal
    (kx, ky) = kernel.shape[:2]
                    #getting number of rows that needed to be pad at the corner of image
    pad = (kx - 1) // 2
                    #using built in feature to pad zero's around the main image
    image = cv2.copyMakeBorder(image, pad, pad, pad, pad, cv2.BORDER_CONSTANT, None, value = 0)
                    #loop for iteration
    for y in np.arange(pad, ix + pad):
            for x in np.arange(pad, iy + pad):                     #extract part of image for convolution through image slicing
                PartOfImageForConvolution = image[y - pad:y + pad + 1, x - pad:x + pad + 1]
                var = (PartOfImageForConvolution * kernel).sum()   #multiplication result
                output[y - pad, x - pad] = var                     #setting result in output array that holds results 
    return output  
  

 



# Calculate fx using Sobel Filters
def fx_(Smoothed_Image):
    fx_mask=np.array([[-1,0,1],
                      [-2,0,2],
                      [-1,0,1]])
    Image=convolution(Smoothed_Image,fx_mask)
    return Image
    
    
# Calculate fy using Sobel Filters
def fy_(Smoothed_Image):
    fy_mask=np.array([[-1,-2,-1],
                      [0,0,0],
                      [1,2,1]])
    Image=convolution(Smoothed_Image,fy_mask)
    return Image
   
def gradient_Magnitude(fx,fy):
    # Make an empty array of image size to hold result 
    gradient = np.zeros(shape=(fx.shape[0], fx.shape[1]))
    (x,y)= fx.shape
                        # loop to get sqrt(fx^2+fy^2) at each position
    for i in range(x):
        for j in range(y):
            gradient[i,j]=np.sqrt(np.power(fx[i,j],2)+np.power(fy[i,j],2))
    return gradient

def gradient_Direction(fx,fy):
    direction=np.arctan2(fy,fx)
    direction=np.rad2deg(direction)
    direction+=180
    return direction

def non_maximum_suppression(gradientMagnitudeImage,angle):
                # image to hold results 
    NonMaximaSuppressedImage = np.zeros(gradientMagnitudeImage.shape)
    imageX=275
    imageY=219
                # loop to create 8 bins and compare current pixel with neighbouring pixels
    for i in range(imageX):
        for j in range(imageY):
            try:
                Neighbour1=255
                Neighbour2=255
                    # Compare with neighbours
                   #angle 0 / first bin
                if (0 <= angle[i,j] < 22.5)  or (157.5 <= angle[i,j] <= 202.5):
                    Neighbour1 = gradientMagnitudeImage[i, j+1]
                    Neighbour2 = gradientMagnitudeImage[i, j-1]
                    #angle 45
                elif (22.5 <= angle[i,j] < 67.5):
                    Neighbour1 = gradientMagnitudeImage[i+1, j-1]
                    Neighbour2 = gradientMagnitudeImage[i-1, j+1]
                    #angle 90
                elif (67.5 <= angle[i,j] < 112.5):
                    Neighbour1 = gradientMagnitudeImage[i+1, j]
                    Neighbour2 = gradientMagnitudeImage[i-1, j]
                    #angle 135
                elif (112.5 <= angle[i,j] < 157.5):
                    Neighbour1 = gradientMagnitudeImage[i-1, j-1]
                    Neighbour2 = gradientMagnitudeImage[i+1, j+1]

                                # If current pixel valuse >Neighbour1 & Neighbour 2 be it same. 
                if (gradientMagnitudeImage[i,j] >= Neighbour1):
                    if(gradientMagnitudeImage[i,j] >= Neighbour2):
                        NonMaximaSuppressedImage[i,j] = gradientMagnitudeImage[i,j]
                else:               # if current pixel value is smaller than both pixels suppress it to zero.
                    NonMaximaSuppressedImage[i,j] = 0
            except IndexError as e:
                pass
    return NonMaximaSuppressedImage



def Thresholding(img,thigh,tlow):
    thresholded=np.zeros((img.shape[0],img.shape[1]))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if(img[i][j]>thigh):
                thresholded[i][j]=255
            elif(img[i][j]<tlow):
                thresholded[i][j]=0
            else:
                thresholded[i][j]=50
    return thresholded


def hysterisis(thresholded_img,original_img):
    hys_thresholded=np.copy(thresholded_img)
    padded_rows=thresholded_img.shape[0]-original_img.shape[0]
    padded_cols=thresholded_img.shape[0]-original_img.shape[0]
    for i in range(int(padded_rows/2),original_img.shape[0]+1):
        for j in range(int(padded_cols/2),original_img.shape[1]+1):
            if(thresholded_img[i][j]==50):
                if(thresholded_img[i-1][j-1]==255 or thresholded_img[i-1][j]==255 or thresholded_img[i-1][j+1]==255 or thresholded_img[i][j-1]==255 or thresholded_img[i][j]==255 or thresholded_img[i][j+1]==255 or thresholded_img[i+1][j-1]==255 or thresholded_img[i+1][j]==255 or thresholded_img[i+1][j+1]==255):
                    hys_thresholded[i][j]=255
    return hys_thresholded 


```

```python
#Step no 1:Read image as gray
image=cv.imread("book.jpg",0)
plt.imshow(image,cmap="gray")



```

```python
guass_filter=np.array([
    [1/16,2/16,1/16],
    [2/16,4/16,2/16],
    [1/16,2/16,1/16]
])
Gausianblurr_img=convolution(image,guass_filter)
plt.imshow(Gausianblurr_img,cmap="gray")

```

```python
fxImage=fx_(Gausianblurr_img)

fyImage=fy_(Gausianblurr_img)

plt.figure(figsize=(20,10))
plt.subplot(141)
plt.imshow(fxImage, cmap='gray')
plt.title('fx')
plt.xticks([])
plt.yticks([])
plt.subplot(142)
plt.imshow(fyImage, cmap='gray')
plt.title('fy')
plt.xticks([])
plt.yticks([])
cv.imwrite("fxImage.jpg",fxImage)
cv.imwrite("fyImage.jpg",fyImage)
```

```python
#find gradient
gradient=gradient_Magnitude(fxImage,fyImage)
plt.imshow(gradient,cmap="gray")
cv.imwrite("Gradient.jpg",gradient)
```

```python
#find gradient direction
direction=gradient_Direction(fxImage,fyImage)
plt.imshow(direction,cmap="gray")
```

```python
#Non Maxima Suppresion:
Suppresed=non_maximum_suppression(gradient,direction)
plt.imshow(Suppresed,cmap="gray")
cv.imwrite("NMS.jpg",Suppresed)
```

```python
#Thresholding

Thresholded_img=Thresholding(Suppresed,120,80)
plt.imshow(Thresholded_img,cmap="gray")
```

```python
#Hysterisis Thresholding
Final=hysterisis(Thresholded_img,image)
plt.imshow(Final,cmap="gray")
cv.imwrite("HysterisisThresholding.jpg",Final)
```
